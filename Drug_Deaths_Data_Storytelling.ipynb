```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Importing the datasets using the paths provided
drug_deaths = pd.read_csv('/Users/arka_bagchi/Desktop/Springboard/Data Storytelling/drug_deaths_storytelling.csv')
drug_deaths.head()
```

```python
# Dropping the redundant column
drug_deaths.drop(columns=['Unnamed: 0'], inplace=True)

# Checking for missing values
missing_values = drug_deaths.isnull().sum()

# Getting a summary of the data
data_summary = drug_deaths.describe(include='all')

missing_values, data_summary
```

```python
# Handling missing values for the demographic information
drug_deaths['Age'].fillna(drug_deaths['Age'].median(), inplace=True)  # Filling age with median
drug_deaths['Sex'].fillna('Unknown', inplace=True)  # Filling sex with 'Unknown'
drug_deaths['Race'].fillna('Unknown', inplace=True)  # Filling race with 'Unknown'

# Handling missing values for location-based data
location_cols = ['ResidenceCity', 'ResidenceCounty', 'ResidenceState', 
                 'InjuryCity', 'InjuryCounty', 'InjuryState']

for col in location_cols:
    drug_deaths[col].fillna('Unknown', inplace=True)

# Handling missing values for other categorical data
drug_deaths['MannerofDeath'].fillna('Unknown', inplace=True)
drug_deaths['DescriptionofInjury'].fillna('Unknown', inplace=True)
drug_deaths['OtherSignifican'].fillna('Unknown', inplace=True)

# Verifying the missing values after the imputations
missing_values_after = drug_deaths.isnull().sum()
missing_values_after
```

```python
# Calculating the correlation matrix for the drug columns
correlation_matrix = drug_deaths[drug_columns].corr()

# Plotting the heatmap for the correlations
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", vmin=-1, vmax=1, square=True)
plt.title('Correlation Matrix of Drugs Detected in Deaths')
plt.show()
```

Fentanyl and Heroin: There's a noticeable negative correlation (-0.43) between Fentanyl and Heroin. This suggests that when Fentanyl is present, Heroin is less likely to be detected, and vice versa. This could indicate that some individuals are primarily using one over the other, rather than combining them.

AnyOpioid: As expected, this has strong positive correlations with individual opioids like Fentanyl, Oxycodone, Methadone, etc., as it's an aggregated measure of any opioid presence.

Given the negative correlation observed between Heroin and Fentanyl, we hypothesize that:

When Fentanyl is present in a drug-related death, Heroin is less likely to be present, and vice versa. This might suggest that individuals are substituting one drug for the other.


To test this hypothesis, we can:

Calculate the proportion of deaths where only Fentanyl is present and Heroin is absent.
Calculate the proportion of deaths where only Heroin is present and Fentanyl is absent.
Compare these proportions to the cases where both are present.

If our hypothesis holds, we should expect the individual presence of each drug (without the other) to be considerably higher than their combined presence.

```python
# Calculating occurrences for different combinations of Heroin and Fentanyl
fentanyl_only = len(drug_deaths[(drug_deaths['Fentanyl'] == 1) & (drug_deaths['Heroin'] == 0)])
heroin_only = len(drug_deaths[(drug_deaths['Fentanyl'] == 0) & (drug_deaths['Heroin'] == 1)])
both_present = len(drug_deaths[(drug_deaths['Fentanyl'] == 1) & (drug_deaths['Heroin'] == 1)])
neither_present = len(drug_deaths[(drug_deaths['Fentanyl'] == 0) & (drug_deaths['Heroin'] == 0)])

# Results
results = {
    'Fentanyl Only': fentanyl_only,
    'Heroin Only': heroin_only,
    'Both Present': both_present,
    'Neither Present': neither_present
}

results
```

This observation supports our hypothesis that individuals might be substituting one drug for the other, leading to deaths where only one of these two drugs is detected.



The distribution of races for deaths with Cocaine detected is different from the overall racial distribution of drug-related deaths.

To test this hypothesis, we'll use the Chi-squared test for goodness of fit. This test compares the observed frequencies (distribution of races for deaths with Cocaine detected) with expected frequencies (overall racial distribution of drug-related deaths).

```python
from scipy.stats import chisquare
```

```python
# Observed frequencies: Distribution of races for deaths with Cocaine detected
observed_frequencies = drug_deaths[drug_deaths['Cocaine'] == 1]['Race'].value_counts()

# Expected frequencies: Overall racial distribution of drug-related deaths
expected_ratios = drug_deaths['Race'].value_counts() / len(drug_deaths)
expected_frequencies = expected_ratios * len(drug_deaths[drug_deaths['Cocaine'] == 1])

# Ensuring that only the races present in both observed and expected frequencies are considered
common_races = observed_frequencies.index.intersection(expected_frequencies.index)
observed_frequencies = observed_frequencies[common_races]
expected_frequencies = expected_frequencies[common_races]

# Adjusting expected frequencies to match the sum of observed frequencies
adjusted_expected_frequencies = expected_frequencies * (observed_frequencies.sum() / expected_frequencies.sum())

# Performing the Chi-squared test for goodness of fit
chi2_statistic, p_value = chisquare(observed_frequencies, adjusted_expected_frequencies)

chi2_statistic, p_value
```

The p-value is extremely small, which suggests that there's a statistically significant difference between the observed and expected racial distributions for deaths with Cocaine detected.

In other words, the racial distribution of deaths where Cocaine was detected is different from the overall racial distribution of drug-related deaths in this dataset. This supports our hypothesis.

```python
# Calculating percentage distributions
observed_percentages = (observed_frequencies / observed_frequencies.sum()) * 100
expected_percentages = (adjusted_expected_frequencies / adjusted_expected_frequencies.sum()) * 100

# Creating a DataFrame for easier plotting
distribution_df = pd.DataFrame({
    'Race': common_races,
    'Observed Percentages': observed_percentages.values,
    'Expected Percentages': expected_percentages.values
})

# Plotting the observed vs expected racial distributions for cocaine deaths
plt.figure(figsize=(14, 8))
bar_width = 0.35
index = np.arange(len(common_races))

plt.bar(index, distribution_df['Observed Percentages'], bar_width, label='Observed', color='b', alpha=0.7)
plt.bar(index + bar_width, distribution_df['Expected Percentages'], bar_width, label='Expected', color='r', alpha=0.7)

plt.xlabel('Race')
plt.ylabel('Percentage (%)')
plt.title('Observed vs Expected Racial Distributions for Cocaine Deaths')
plt.xticks(index + bar_width / 2, distribution_df['Race'], rotation=45)
plt.legend()
plt.tight_layout()
plt.grid(axis='y', linestyle='--', linewidth=0.5)
plt.show()
```

The bar plot provides a visual comparison of the observed racial distribution (for deaths with Cocaine detected) versus the expected racial distribution (based on the overall dataset):

White Individuals: The observed percentage of White individuals in Cocaine-related deaths is lower than the expected percentage.
Black Individuals: A significant discrepancy is observed for Black individuals, where the percentage is much higher in Cocaine-related deaths compared to the expected distribution.
Hispanic Individuals: The observed and expected percentages are relatively close, with a slightly higher observed percentage for Cocaine-related deaths.
Other racial categories (Asian/Pacific Islander, Native American, and others) have relatively smaller differences between observed and expected distributions.
The most notable difference is for Black individuals, indicating a potential disproportionate impact of Cocaine on this racial group compared to the overall dataset.
