```python
# First let's check the shape of the DataFrame
print("The dataset has {} rows and {} columns.".format(properties.shape[0], properties.shape[1]))

# Then let's see the first few rows of the data
print(properties.head())

# Now let's see the data types of each column
print(properties.dtypes)

# Finally, let's see if there are any missing values
print(properties.isnull().sum())
```

**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

Don't be afraid to use StackOverflow for help  with this.

```python
# Transpose the DataFrame and reset the index
properties_T = properties.transpose().reset_index()

# Assign the values of the first row to the column headers
properties_T.columns = properties_T.iloc[0]

# Drop the first row
properties_T = properties_T.drop(0)

# Display the DataFrame
properties_T.head()
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
# Rename the columns
properties_T = properties_T.rename(columns = {'Unnamed: 0':'London_Borough', pd.NaT: 'ID'})

# Check the DataFrame
properties_T.head()
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
# Use the melt function to reshape the DataFrame, moving the monthly dates into a single column and the prices along with them
properties_melt = pd.melt(properties_T, id_vars= ['London_Borough', 'ID'])

# Rename the columns
properties_melt = properties_melt.rename(columns = {0: 'Month', 'value': 'Average_price'})

# Display the DataFrame
properties_melt.head()
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
# Check the data types of each column
print(properties_melt.dtypes)

# Convert 'Average_price' to a numeric type (float)
properties_melt['Average_price'] = pd.to_numeric(properties_melt['Average_price'])

# Check the data types of each column again
print(properties_melt.dtypes)
```

**2.5. Cleaning the data (part 3)**

Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit. 

```python
# Check unique 'London_Borough' values
print(properties_melt['London_Borough'].unique())

# A list of all the Boroughs
boroughs = ['City of London', 'Barking & Dagenham', 'Barnet', 'Bexley', 'Brent', 'Bromley', 'Camden', 'Croydon', 'Ealing', 
            'Enfield', 'Greenwich', 'Hackney', 'Hammersmith & Fulham', 'Haringey', 'Harrow', 'Havering', 'Hillingdon', 
            'Hounslow', 'Islington', 'Kensington & Chelsea', 'Kingston upon Thames', 'Lambeth', 'Lewisham', 'Merton', 
            'Newham', 'Redbridge', 'Richmond upon Thames', 'Southwark', 'Sutton', 'Tower Hamlets', 'Waltham Forest', 
            'Wandsworth', 'Westminster']

# Only keep rows whose 'London_Borough' is in the boroughs list
properties_melt = properties_melt[properties_melt['London_Borough'].isin(boroughs)]

# Check the number of rows again
print(properties_melt.count())

# Check for missing values
print(properties_melt.isnull().sum())
```

**2.6. Visualizing the data**

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

```python


# Subset the data for Islington
islington_prices = properties_melt[properties_melt['London_Borough'] == 'Islington']

# Create a plot of Month against Average_price
plt.figure(figsize=(10,5))
plt.plot(islington_prices['Month'], islington_prices['Average_price'])

# Add title and labels
plt.title('Average House Prices in Islington Over Time')
plt.xlabel('Month')
plt.ylabel('Average Price')

# Display the plot
plt.show()
```

To limit the number of data points you have, you might want to extract the year from every month value your *Month* column.

```python
# Extract the year from each month value in the 'Month' column
properties_melt['Year'] = properties_melt['Month'].apply(lambda t: t.year)

# Check the DataFrame
properties_melt.head()
```

```python
# Group by 'London_Borough' and 'Year', and calculate the mean
dfg = properties_melt.groupby(by=['London_Borough', 'Year']).mean()

# Display a sample of the DataFrame
dfg.sample(10)

# Reset the index
dfg = dfg.reset_index()

# Display the first few rows of the DataFrame
dfg.head()
```

```python
# Calculate the mean price for 1998 and 2023
avg_price_1998 = dfg[dfg['Year'] == 1998].groupby('London_Borough').mean()['Average_price']
avg_price_2023 = dfg[dfg['Year'] == 2023].groupby('London_Borough').mean()['Average_price']

# Calculate the price increase for each borough
price_increase = avg_price_2023 - avg_price_1998

# Sort the boroughs by price increase and select the top 5
top5_boroughs = price_increase.sort_values(ascending=False).head(5)

# Display the top 5 boroughs
print(top5_boroughs)

#Kensington & Chelsea, City of London, Westminster, Camden, and Richmond upon Thames have all had the highest mean price increase
#This is the pure notional value of the houses going up, not reflecting the proportion of growth 
```

```python
# Sort the DataFrame by 'Average_price' in descending order
dfg_sorted = dfg.sort_values(by='Average_price', ascending=False)

# Display the first few rows of the sorted DataFrame
dfg_sorted.head()

#This shows us that recently in 2022, Kensington and Chelesa has the highest average price. 
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
def create_price_ratio(d):
    """Calculate the ratio of average house prices between 1998 and 2018."""
    price1998 = float(d['Average_price'][d['Year']==1998])
    price2018 = float(d['Average_price'][d['Year']==2018])
    ratio = [price2018/price1998]
    return ratio

# Test the function with one borough
print(create_price_ratio(dfg[dfg['London_Borough']=='Camden']))
```

```python
# Which boroughs have top 5 price ratios? Which boroughs have bottom 5 price ratios?
# Initialize an empty dictionary to store the results
ratios = {}

# Calculate the ratio for each borough and store it in the dictionary
for b in dfg['London_Borough'].unique():
    borough = dfg[dfg['London_Borough'] == b]
    ratios[b] = create_price_ratio(borough)

# Convert the dictionary to a DataFrame
df_ratios = pd.DataFrame(ratios).transpose().reset_index()

# Rename the columns
df_ratios.rename(columns={'index':'Borough', 0:'2018/1998 ratio'}, inplace=True)

# Sort by ratio and display the top 5 boroughs
top5 = df_ratios.sort_values(by='2018/1998 ratio',ascending=False).head(5)
print("Top 5 boroughs with the highest price increase ratio:")
print(top5)

# Sort by ratio and display the bottom 5 boroughs
bottom5 = df_ratios.sort_values(by='2018/1998 ratio',ascending=True).head(5)
print("\nTop 5 boroughs with the lowest price increase ratio:")
print(bottom5)
```
